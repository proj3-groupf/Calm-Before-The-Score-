```python
import numpy as np
import pandas as pd
import requests
from pprint import pprint
import pymongo
```

```python
#base url for API call to sportsdata.io
base_url = "https://api.sportsdata.io/v3/nfl/scores/json/Scores/2021?key=ec966d78fab6468eaa542e1e7e883a44"
```

```python
#test response for connection
response = requests.get(base_url)
print(response)
```

```python
#pretty print JSON return
scores2021 = response.json()
# pprint(scores2021)
```

```python
#Create DataFrame from JSON
scores2021DF = pd.DataFrame(scores2021)
scores2021DF
```

```python
scores2021DF.columns
```

```python
scores2021cleanDF = scores2021DF.drop(columns=['StadiumDetails'])
```

```python
#api call for stadium data from sportsdata.io
stadium_url = "https://api.sportsdata.io/v3/nfl/scores/json/Stadiums?key=ec966d78fab6468eaa542e1e7e883a44"
```

```python
#test response for connection
stadium_response = requests.get(stadium_url)
print(stadium_response)
```

```python
#JSON and pretty print of stadium response
stadiums = stadium_response.json()
pprint(stadiums)
```

```python
stadiumDF = pd.DataFrame(stadiums)
stadiumDF
```

```python
mergedDF = scores2021cleanDF.merge(stadiumDF, on='StadiumID')
mergedDF
```

```python
# create MongoDB connection using pymongo 

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
db = client.nfl_db

# Declare the collection
season_data = db.season_data
```

```python
season_data.insert_many(mergedDF.to_dict('records'))
```
