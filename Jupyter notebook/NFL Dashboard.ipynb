```python
import numpy as np
import pandas as pd
import requests
from pprint import pprint
import pymongo
```

```python
#base url for API call to sportsdata.io
base_url = "https://api.sportsdata.io/v3/nfl/scores/json/Scores/2021?key=ec966d78fab6468eaa542e1e7e883a44"
```

```python
#test response for connection
response = requests.get(base_url)
print(response)
```

```python
#pretty print JSON return
scores2021 = response.json()
pprint(scores2021)
```

```python
#Create DataFrame from JSON
scores2021DF = pd.DataFrame(scores2021)
scores2021DF
```

```python
scores2021DF.columns
```

```python
scores2021cleanDF = scores2021DF.drop(columns=['StadiumDetails','PointSpread',
       'OverUnder', 'Quarter', 'TimeRemaining', 'Possession', 'Down',
       'Distance', 'YardLine', 'YardLineTerritory', 'RedZone',
       'AwayScoreQuarter1', 'AwayScoreQuarter2', 'AwayScoreQuarter3',
       'AwayScoreQuarter4', 'AwayScoreOvertime', 'HomeScoreQuarter1',
       'HomeScoreQuarter2', 'HomeScoreQuarter3', 'HomeScoreQuarter4',
       'HomeScoreOvertime', 'HasStarted', 'IsInProgress', 'IsOver',
       'Has1stQuarterStarted', 'Has2ndQuarterStarted', 'Has3rdQuarterStarted',
       'Has4thQuarterStarted', 'IsOvertime', 'DownAndDistance',
       'QuarterDescription', 'LastUpdated', 'GeoLat', 'GeoLong','GlobalAwayTeamID', 'GlobalHomeTeamID', 'PointSpreadAwayTeamMoneyLine',
       'PointSpreadHomeTeamMoneyLine', 'ScoreID', 'Status', 'GameEndDateTime',
       'HomeRotationNumber', 'AwayRotationNumber', 'NeutralVenue', 'RefereeID',
       'OverPayout', 'UnderPayout', 'HomeTimeouts', 'AwayTimeouts',
       'DateTimeUTC',])
```

```python
#api call for stadium data from sportsdata.io
stadium_url = "https://api.sportsdata.io/v3/nfl/scores/json/Stadiums?key=ec966d78fab6468eaa542e1e7e883a44"
```

```python
#test response for connection
stadium_response = requests.get(stadium_url)
print(stadium_response)
```

```python
#JSON and pretty print of stadium response
stadiums = stadium_response.json()
pprint(stadiums)
```

```python
stadiumDF = pd.DataFrame(stadiums)
stadiumDF
```

```python
mergedDF = scores2021cleanDF.merge(stadiumDF, on='StadiumID')
mergedDF
```

```python
team_url = "https://api.sportsdata.io/v3/nfl/scores/json/Teams/2021?key=ec966d78fab6468eaa542e1e7e883a44"
#test response for connection
team_response = requests.get(team_url)
print(team_response)
```

```python
#JSON and pretty print of team response
teams = team_response.json()
teamsDF = pd.DataFrame(teams)
teamsCleanDF = teamsDF.drop(columns=['ByeWeek', 'AverageDraftPosition',
       'AverageDraftPositionPPR', 'UpcomingSalary', 'UpcomingOpponent',
       'UpcomingOpponentRank', 'UpcomingOpponentPositionRank',
       'UpcomingFanDuelSalary', 'UpcomingDraftKingsSalary',
       'UpcomingYahooSalary', 'DraftKingsName',
       'DraftKingsPlayerID', 'FanDuelName', 'FanDuelPlayerID',
       'FantasyDraftName', 'FantasyDraftPlayerID', 'YahooName',
       'YahooPlayerID', 'AverageDraftPosition2QB',
       'AverageDraftPositionDynasty', 'StadiumDetails'])

teamsfinalDF = teamsCleanDF.rename(columns={"Key": "HomeTeam"})
teamsfinalDF
```

```python
nflDataFinalDF = mergedDF.merge(teamsfinalDF, on='HomeTeam')
nflDataFinalDF
```

```python
# create MongoDB connection using pymongo 

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Declare the database
db = client.nfl_db

# Declare the collection
season_data = db.season_data
```

```python
season_data.insert_many(nflDataFinalDF.to_dict('records'))
```
